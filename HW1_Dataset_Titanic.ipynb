```python
means_ages_group1 = []
means_ages_group2 = []

n = 1
while n < 891:
    n += 1
    sample1 = df[(df["Sex"] == "male")]["Age"].dropna().sample(50).mean()
    sample2 = df[(df["Sex"] == "female")]["Age"].dropna().sample(50).mean()

    means_ages_group1.append(sample1)
    means_ages_group2.append(sample2)
```

```python
from scipy.stats import shapiro
from scipy.stats import levene
```

```python
shapiro(means_ages_group1)
```

```python
shapiro(means_ages_group2)
```

```python
age_survived = df[(df['Sex'] == "male")]['Age'].dropna()
age_died = df[(df['Sex'] == "female")]['Age'].dropna()
```

```python
levene(age_survived, age_died)
```

```python
from scipy.stats import ttest_ind
ttest_ind(age_survived, age_died, equal_var = False)
```

```python
print(age_survived.mean())
print(age_died.mean())
```
